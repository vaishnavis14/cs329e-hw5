```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
from astropy.coordinates import SkyCoord
from astropy import units
```

# Q1

```python
df = pd.read_csv('bsc5.csv')
df.head()
```

# Q2

```python
df = df[df['Parallax'].isnull() == False]
df.head()
```

# Q3

```python
df['Distance'] = 1 / df['Parallax']
df.head()
```

# Q4

```python
def right_ascension(hours, mins, secs):
  return (hours + (mins / 60) + (secs / 3600)) * 15
df['ra'] = right_ascension(df['RAh'], df['RAm'], df['RAs'])
def declination(degrees, mins, secs):
  return degrees + (mins / 60) + (secs / 3600)
df['dec'] = declination(df['DEd'], df['DEm'], df['DEs'])
df.head()
```

# Q5

```python
spherical_coord = SkyCoord(ra = df['ra'], dec = df['dec'], distance = df['Distance'], unit=(units.degree, units.degree))
df['x'] = spherical_coord.cartesian.x
df['y'] = spherical_coord.cartesian.y
df['z'] = spherical_coord.cartesian.z
df.head()
```

# Q6

```python
df['color'] = np.nan
for i in df['SpType'].index:
  letter = str(df['SpType'][i]).strip()[0].upper()
  if letter == 'O':
    df['color'][i] = 'deepestblue'
  elif letter == 'B':
    df['color'][i] = 'mediumblue'
  elif letter == 'A':
    df['color'][i] = 'lightblue'
  elif letter == 'F':
    df['color'][i] = 'green'
  elif letter == 'G':
    df['color'][i] = 'yellow'
  elif letter == 'K':
    df['color'][i] = 'orange'
  elif letter == 'M':
    df['color'][i] = 'red'
  else:
    df['color'][i] = 'black'
df.head()
```

# Q7, Q8, Q9

```python
color_discrete_map = {'deepestblue': 'rgb(5, 1, 74)', 'mediumblue': 'rgb(0,130,230)', 'lightblue': 'rgb(120,206,255)',
                      'green' : 'rgb(0,128,0)', 'yellow' : 'rgb(255, 204, 0)', 'orange' : 'rgb(255, 165, 0)',
                      'red' : 'rgb(139,0,0)', 'black' : 'rgb(0,0,0)'}
fig = px.scatter_3d(df, 'x', 'y', 'z', df['color'], color_discrete_map = color_discrete_map, hover_data = ['HR', 'Distance', 'RadVel'])
fig.update_traces(marker=dict(size=3))
fig.update_layout(
    title='3D Distribution of Stars',
    scene=dict(xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'))
fig.show()
```
